Africa geospatial visualization

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
dati = pd.read_excel(r"C:\Users\39340\Desktop\Data-Visualization\Africa child mortality\africa_data.xlsx")

dati_subset = dati[['Country Code','Series Name', 'Series Code', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']]

columns_name = {
    '2017 [YR2017]': "2017",
    '2018 [YR2018]' : "2018",
    '2019 [YR2019]' : "2019",
    '2020 [YR2020]' : "2020",
    '2021 [YR2021]': "2021"
}

dati_subset = dati_subset.rename(columns=columns_name)

dati_subset
```

```python
import math

dati_db = dati_subset[dati_subset["Series Name"] == "Completeness of birth registration (%)"].reset_index()

dati_db.replace("..", math.nan, inplace=True)

dati_db
```

```python
world_df = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_110m_admin_0_countries.geojson')

world_df
```

```python
dati_merge = dati_db.merge(world_df, left_on=dati_db["Country Code"], right_on=world_df["adm0_a3"])

geo_db = gpd.GeoDataFrame(dati_merge, geometry="geometry")
```

```python
africa_bounds = {
    'minx': -2500000,  # west
    'miny': -4000000,  # south
    'maxx': 6000000,   # east
    'maxy': 4000000    # north
}

geo_db = geo_db.to_crs(epsg=3857)

for i in ["2017", "2018", "2019", "2020", "2021"]:
    ax = geo_db.plot(i, cmap='viridis', missing_kwds = {"color": "#d2d9d4", "edgecolor": "white",
            "hatch": "///"}, legend = True)
    ax.set_xlim(africa_bounds['minx'], africa_bounds['maxx'])
    ax.set_ylim(africa_bounds['miny'], africa_bounds['maxy'])
    ax.set_axis_off()
```
